# GOALS 3.7 AND 3.8

### 3.7 By 2030, ensure universal access to sexual and reproductive health-care services, including for family planning, information and education, and the integration of reproductive health into national strategies and programmes


By 2030, the goal is to ensure that everyone has access to comprehensive sexual and reproductive healthcare services, encompassing family planning, education, and the seamless integration of reproductive health into national programs. This initiative aims to promote informed choices and well-being for individuals worldwide.

3.7.1 Proportion of women of reproductive age (aged 15–49 years) who have their need for family planning satisfied with modern methods

3.7.2 Adolescent birth rate (aged 10–14 years; aged 15–19 years) per 1,000 women in that age group

### Purpose 


Goal 3.7 of the Sustainable Development Goals (SDGs) is a critical component of the broader agenda to improve global health and well-being. The goal, titled "Ensure universal access to sexual and reproductive health-care services, including for family planning, information and education, and the integration of reproductive health into national strategies and programmes," has profound implications for individual and societal health, gender equality, economic development, and social justice.

Family planning is a key element within Goal 3.7, acknowledging the right of individuals and couples to decide the timing and spacing of their children. Access to family planning methods empowers individuals to make informed choices about their reproductive health, which in turn positively impacts their overall well-being. By providing access to a variety of contraceptive methods, education, and counseling, societies can help individuals plan and have the number of children they desire, ultimately contributing to healthier families and communities.

Moreover, the integration of reproductive health into national strategies and programs underscores the need to mainstream reproductive health within broader health frameworks. This integration ensures that reproductive health concerns are not isolated but are interconnected with overall health policies, enabling a more holistic and effective approach to healthcare. It emphasizes the importance of addressing reproductive health needs at all stages of life, from adolescence to adulthood and into the later years.

### 3.8 Achieve universal health coverage, including financial risk protection, access to quality essential health-care services and access to safe, effective, quality and affordable essential medicines and vaccines for all

Goal 3.8 aims to guarantee that everyone has access to quality healthcare services without suffering financial hardship, encompassing essential medicines and vaccines, fostering a healthier and more equitable global population. This inclusivity ensures the provision of safe and effective healthcare, enhancing overall public health.

3.8.1 Coverage of essential health services (defined as the average coverage of essential services based on tracer interventions that include reproductive, maternal, newborn and child health, infectious diseases, non-communicable diseases and service capacity and access, among the general and the most disadvantaged population)

3.8.2 Proportion of population with large household expenditures on health as a share of total household expenditure or income


### Purpose

Goal 3.8 of the Sustainable Development Goals (SDGs) focuses on achieving universal health coverage, ensuring financial risk protection, access to quality essential health-care services, and accessibility to safe, effective, quality, and affordable essential medicines and vaccines for all. This goal holds immense significance in addressing global health challenges and fostering a world where health is a fundamental right accessible to everyone.

Financial risk protection is a key element of this goal, aiming to shield individuals and families from the financial burden associated with healthcare. High medical costs can push people into poverty or prevent them from seeking timely medical attention. Goal 3.8 envisions a world where individuals are not forced to choose between their health and financial stability, ensuring that health expenses do not lead to impoverishment.

Access to quality essential health-care services is another vital aspect. This encompasses a wide range of health services, including preventive, diagnostic, treatment, and rehabilitative services. The goal emphasizes the importance of providing healthcare services that meet high standards of quality, thereby ensuring optimal health outcomes for individuals and communities.

The goal also stresses the importance of accessible, safe, effective, and affordable essential medicines and vaccines. Medicines and vaccines are fundamental components of healthcare, aiding in the prevention, treatment, and control of diseases. Ensuring their accessibility and affordability is crucial to improving public health, controlling disease outbreaks, and enhancing overall well-being.
